### Exploring the features with respect to `diagnosis` 

```python
target_df = data_df.drop(['id'], axis=1)
x = target_df.drop(['diagnosis'], axis=1)
y = target_df.diagnosis # M or B
plot.feature_vs_target(target_df, x)
```

### Correlation Between Features

```python
plot.heatmap(x, "Correlation between features")
```

### Outlier Treatment

```python
outlier = DfOutlier(x)
outlier.getOverview()
```

```python
# Normalizing the features
normalized_features = (x - x.mean()) / (x.std())
```

```python
plot.multiple_boxplot(normalized_features, y, 0, 8)
```

```python
plot.multiple_boxplot(normalized_features, y, 9, 18)
```

```python
plot.multiple_boxplot(normalized_features, y, 18, 31)
```
